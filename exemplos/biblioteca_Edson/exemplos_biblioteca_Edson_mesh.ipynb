Malha setando Rho e calculando por Rho (FlagRhoBased == True)

Malha setando Sigma mas calculando por Rho

```python
M4 = mesh.PointElectrodes2DMeshEdson(8, nome_msh=nome, altura2D = 0.02)
M4.ReadMesh()
meus_sigmas = {}
meus_sigmas[1] = 0.1
meus_sigmas[2] = 0.2
M4.SetSigmaPhysicaEntity(meus_sigmas) # Informando sigma (e já calculando o rho de cada elemento)
M4.FlagRhoBased = True
M4.CalcKGlobal() # calculando KGlobal usando Rho
M4.KGlobal
```

Ajustando Sigma por vetor (tamanho nro de elementos), e não por Physical Entity

```python
Mvec = mesh.PointElectrodes2DMeshEdson(8, nome_msh=nome, altura2D = 0.02)
Mvec.ReadMesh()
vec = np.ones(Mvec.NumberOfElements) * 0.1
Mvec.SetSigmaElements(vec) # Informando sigma (e já calculando o rho de cada elemento)
Mvec.FlagRhoBased = True
Mvec.CalcKGlobal() # calculando KGlobal usando Rho
Mvec.KGlobal
```

# Comparação de elementos 2D Edson vs Erick

```python
nome = '../../malhasMSH/circ16_anomalia6.msh'
MinhaMalhaEdson = mesh.PointElectrodes2DMeshEdson(16, nome_msh=nome, altura2D = 0.02, useEdson=True)
MinhaMalhaErick = mesh.PointElectrodes2DMeshEdson(16, nome_msh=nome, altura2D = 0.02, useEdson=False)

MinhaMalhaEdson.ReadMesh()
MinhaMalhaErick.ReadMesh()
```

```python
MinhaMalhaEdson.MshFileName
```

```python
MinhaMalhaEdson.SetRhoValue(10)
MinhaMalhaEdson.CalcKGlobal()

MinhaMalhaErick.SetRhoValue(10)
MinhaMalhaErick.CalcKGlobal()

for linha in range(MinhaMalhaEdson.NumberOfNodes):
    dif =  np.max(np.abs(MinhaMalhaEdson.KGlobal[linha]-MinhaMalhaErick.KGlobal[linha]))
    if dif > 1e-7:
        print(f'linha {linha}: {dif}')

print(f'Erro maximo: {np.max(np.abs(MinhaMalhaEdson.KGlobal-MinhaMalhaErick.KGlobal))}')
```
